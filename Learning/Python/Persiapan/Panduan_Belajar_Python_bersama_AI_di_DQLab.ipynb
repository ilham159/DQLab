**Kenapa Belajar Data harus Coding?**

Ini adalah pertanyaan yang sering diajukan kepada DQLab, kenapa tidak kemampuan *spreadsheet* atau Excel saja yang sudah banyak fiturnya?

Jawaban kami adalah keterampilan *coding* sangat penting untuk profesi data, baik *data analyst*, *data engineer* maupun *data scientist* karena mereka biasanya bekerja dengan data dengan volume atau variasi yang besar dan kompleks.

Dengan demikian para profesi tersebut memerlukan *tool* dan teknik khusus untuk mengelola, menganalisa, dan menghasilkan kesimpulan dan fenomena dari data. Hal ini seringkali memerlukan keterampilan *coding* dengan menggunakan bahasa seperti Python, R, atau SQL untuk melakukan hal-hal tersebut.

Berikut adalah beberapa alasan mengapa keterampilan coding sangat penting bagi kita:

*   **Transformasi Data**: Ini terutama untuk data engineer dimana profesi ini perlu membersihkan dan merubah struktur isi dan format data sebelum dapat dianalisa lebih lanjut. Data scientist dan data analyst sering sekali juga perlu menguasai kemampuan ini. 
*   **Analisis Data**: Setelah data telah dibersihkan dan dibangun strukturnya, data scientist dan data analyst perlu menggunakan model statistik dan algoritma *machine learning* untuk menganalisis data dan menghasilkan informasi bernilai tinggi. Algoritma-algoritma yang digunakan ini sering memerlukan keterampilan *coding* untuk menerapkannya dan mengoptimalkannya.
*   **Otomatisasi**: Analis data dan ilmuwan data sering bekerja dengan dataset yang besar yang memerlukan tugas yang repetitif seperti *data cleansing* atau pembersihan data, visualisasi data, dan membuat laporan. Keterampilan coding memungkinkan para profesional ini untuk mengotomatisasi tugas-tugas ini menggunakan script yang dibuat dengan bahasa pemrograman tertentu, menghemat waktu mereka untuk analisis dan pengambilan keputusan yang lebih kompleks.
*   **Kolaborasi**: Analis data dan ilmuwan data bekerja dalam tim dengan profesional lain, seperti developer, ui/ux/visual designer, dan project manager. Keterampilan coding memungkinkan mereka untuk berkolaborasi secara efektif dengan tim ini, karena mereka dapat berkomunikasi menggunakan bahasa pemrograman maupun tool umum.

Kesimpulannya adalah keterampilan coding sangat penting bagi para professional data karena memungkinkan mereka untuk mentransformasi dan menganalisis data dengan efektif, mengotomatisasi tugas yang repetitif, dan berkolaborasi secara efektif dengan profesional lainnya.

**Mengenal Live Code Editor**

**Live Code Editor (LCE)** adalah fitur belajar yang sedang Anda gunakan saat ini. Fitur ini berbeda dengan bagian sebelumnya. Tujuan fitur ini adalah untuk menunjang pengalaman belajar yang baik dengan cara praktik coding dan mendapatkan feedback langsung. LCE sendiri adalah fitur yang sangat lengkap agar Anda mendapatkan pengalaman belajar *coding* yang terbaik, namun tetap sederhana. Hal ini bertujuan untuk memudahkan pemula dalam mengikutinya, sementara juga tetap nyaman digunakan oleh yang sudah berpengalaman.

LCE terdiri dari tiga panel utama:

Mengenal Live Code Editor.png

*   **Panel Materi**
Panel materi ini terdiri dari dua tab, yaitu **Theory** dan **Hint**. Tab Theory berisi teks yang memberikan penjelasan tentang topik yang sedang difokuskan dan juga menyediakan soal praktek. Sementara itu, bagian Hint berisi petunjuk penyelesaian soal praktek jika Anda menekan tombol Hint. Tombol tersebut terletak di bawah LCE.

Panel ini merupakan sebuah *editor* di mana kita dapat memasukkan kode kita. Untuk bekerja dengan kode ini, panel ini memiliki dua tombol, yaitu tombol **Run** untuk menjalankan atau mengeksekusi kode dan tombol **Submit** untuk mengirim kode ini ke sistem penilaian DQLab.

**Code-Based**

Dalam evaluasi praktek coding, DQLab memiliki dua pendekatan, yaitu **code-based** dan **result-based**.

Pendekatan code-based mengacu pada evaluasi berdasarkan kode yang dikirimkan ke sistem DQLab, sedangkan pendekatan result-based mengacu pada evaluasi berdasarkan hasil eksekusi kode yang dikirimkan ke sistem DQLab.

Untuk evaluasi code-based maka kode yang dikirimkan harus sama dengan yang diinstruksikan atau diharapkan. Biasanya ini jelas dalam instruksi maupun hint.

Cara evaluasi code-based ini ada pada tipe Live Module, yaitu modul belajar yang memiliki Live Code Editor. Saat ini topik belajar yang memiliki Live Module adalah Python, SQL, dan R.

Sebagai contoh untuk kode Python yang harus dikirimkan dan diinginkan oleh sistem sesuai instruksi adalah sebagai berikut

    print("Haloooo" + "o")
kemudian Anda kirimkan code 

    print("Haloooo" + o)
Perhatikan kalau terdapat perbedaan tanda kutip pada huruf o sebelum penutupan kurung, maka kode akan dianggap salah oleh sistem DQLab.

Contoh lainnya adalah sebagai berikut

    print(abs(4.0))
kemudian Anda kirimkan code 

    print(fabs(4.0))
Walaupun keduanya jalan, tapi **fabs** bukan diharapkan oleh sistem sehingga akan salah.

Tipe evaluasi code ini sangat cocok untuk pemula yang perlu mengikuti persis apa yang diberikan oleh materi, sehingga bisa menjadi bekal untuk mengerti apa yang diarahkan materi selanjutnya.

Klik tombol **Next** untuk melanjutkan.

**Result-Based**

Untuk evaluasi result-based maka materi berfokus kepada output yang diharapkan. Bentuk code dengan cara dan algoritma bebas digunakan, yang penting dapat menghasilkan output yang diinginkan.

Cara evaluasi result-based ini ada pada tipe Live Project, yaitu project-project yang memberikan kasus untuk dipecahkan dengan menggunakan pengiriman dengan Live Code Editor. Saat ini topik yang memiliki Live Project adalah Python, SQL, dan R.

Sebagai contoh output yang harus dipecahkan untuk code Python adalah sebagai berikut:

```
Memiliki dua input variable berisi angka 3 dan 5, dan mencetak angka 8 yang merupakan hasil penjumlahan kedua angka tersebut. 
```

Code yang harus dikirimkan itu bisa saja banyak variasi, DQLab memberikan dua contoh dari banyak variasi tersebut di bawah ini dimana keduanya diterima oleh sistem DQLab.

Variasi pertama adalah sebagai berikut

```
num1 = 3
num2 = 5
print(num1 + num2)
```
dan variasi kedua adalah sebagai berikut

```
a = 3
b = 5
hasil = a + b
print(hasil)
```

Keduanya jalan dan diterima oleh sistem DQLab. Tipe evaluasi output ini sangat cocok untuk yang sudah cukup matang belajar dan ingin menguji pengetahuannya dengan melihat kebutuhan output yang diminta. Cara evaluasi ini mendekati kenyataan sehari-hari di pekerjaan professional dimana manajemen selalu minta hasil dan caranya diberikan kepada kita.

Klik tombol **Next** untuk melanjutkan.
